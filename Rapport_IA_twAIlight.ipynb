## <a name="alg"></a>3. Algorithmes implémentés

Ces algorithmes reposent sur une surcharge de la fonction next_moves du joueur (JoueurInterne ou Joueur) qui tend généralement à simplifier et à guider de façon plus intelligente l'exploration de notre arbre.

Remarque: ces méthodes ne sont généralement pas exclusives et reposent sur des améliorations apportées à différentes étapes du choix du next_moves de nos unités. Autrement dit nous pouvons les combiner (par exemple: Negamax avec transposition et most probable outcome)

### <a name="subalg1"></a>A - Algorithmes

- #### MinMax

Réécriture de la fonction next_moves qui applique l'algorithme MinMax au choix de la position.

- #### Variante Negamax

Simplification du MinMax qui repose sur le fait que nous sommes dans un jeu à somme nulle.

- #### Variante Negascout

Amélioration du Negamax n'explorant que des noeuds n'étant pas élagués par l'alphabéta.

- #### MonteCarloTreeSearch

Exploration de l'arbre de jeu, selon un critère statistique (Upper Confidence Bound), par simulations. 

Nous avons reussi sur nos ordinateurs avec les approximations décrites plus tard (MPOO) à réaliser 3000 simulations en 2s. Nous utilisons la table de transposition pour enregistrer les simulations d'une carte de départ. Cette approche bien que prometteuse grâce à sa granularité de calcul, ne marche pas aussi bien en pratique. Certaines simulations peuvent durer tellement de temps que l'algorithme peut sauter son tour facilement.

### <a name="subalg2"></a>B - Optimisations

* #### Ajout d'une table de hashage
  Cela permet de garder en mémoire les scores des cartes déjà rencontrées afin de ne pas les recalculer (économie de calcul). Tout comme pour le jeu du Morpion, nous avons repris la table de hashage de Zobrist pour hasher les positions d'une carte.


* #### Réduction de la quantité de données copiées à chaque noeuds
  Chaque noeud copie moins de données vers ses noeuds fils.
  
* #### Choix de la structure de données
  Nous avons essayé plusieurs manières de stocker le contenu de la carte afin de gagner en temps de calcul :
  - Notre premier choix fut de stocker le contenu de la carte sous la forme d'un **dictionnaire python "plein"**. Le dictionnaire était initialisé en remplissant toutes les valeurs de (0,0,0).
      ```
        map.content = {(x, y): (n_hum, n_vamp, n_lg) for each (x,y)}
      ```
  Cette structure très simple s'est avérée être la plus polyvalente de celles que nous avons testées.
  
  - Nous avons essayé de faire une variante **"matrice creuse"** en ne conservant que les cases non vides dans le dictionnaire. Le gain obtenu pour parcourir le dictionnaire (plus petit) était compensé par le temps supplémentaire que prenait la suppression d'un couple clé:valeur par rapport à une modification.
      ```
        map.content = {(x, y): (n_hum, n_vamp, n_lg) for each (x,y) if not (x,y) is empty}
      ```
  
  - Nous avons essayé d'implémenter une version réalisée avec **numpy** afin de profiter de la rapidité de ce module. Le contenu de la carte était alors un **numpy array de forme (largeur_carte, hauteur_carte, 3)**. Le 3 correspond aux triplets humain, vampires, loup-garou.
  Les gains très importants sur toutes les opérations de cette structure de données ont été annulé par le temps de création d'un numpy array très supérieur à celui d'un dict python. Si au lieu de copier le contenu de la carte à chaque création d'un nouveau noeud, nous avions tenu une liste d'update à réaliser, nous aurions probablement obtenu des performances bien meilleures.

### <a name="subalg3"></a>C - Réduction du facteur de branchement

* #### Simplification de la prise en compte des probabilités: Most Probable Outcome
  Jusqu'ici nous créions un noeud pour chaque tirage possible de chaque combat. Afin de réduire le nombre de noeuds générés dans nos arbres, nous avons décidé de ne considérer que l'issue du combat la plus probable (correspondant à l'espérance définie par la loi de probabilité des combats).<br>
  Cette variantes est nommée "**Most Probable Outcome**" et les fichiers l'implémentant comporte "**MPO**" dans leur noms


* #### Réduction du nombre de mouvements explorés à chaque profondeur de l'arbre
  Nous avons commencé par générer **tous** les mouvements possibles à chaque tour. Nous les générions en faisant la produit cartésien des possibilités de chaque groupe. 
  L'explosion combinatoire que ce choix implique dès que la carte prend en complexité ou que les groupes grossissent nous a obligé à revoir notre approche. Nous avons commencé par limité le nombre de mouvements de possibilités pour chaque groupes : <br>
  
  - Réduction du nombre de cases proposées à chaque groupe : <br>
  Au lieu de proposer à chaque groupe de monstres les (au plus) 8 cases l'environnant, on choisit un nombre de cases fixés (passé en paramêtre au JoueurInterne). Il est important de noter que cette mesure ne nous permet pas de contrôler strictement le facteur de branchement mais de le réduire fortement. <br>
  Ces cases sont choisies selon une heuristique. Plusieurs essais ont été réalisés, des heuristiques les plus simplistes (nombre d'humains sur la case) à des heuristiques beaucoup plus lourde comprenant des produits de convolution de la carte pour extraire des informations supplémentaires. <br> 
  Les versions simples (ordonnant les cases uniquement en fonction de leur contenus) étaient très rapides nous permettant de calculer plus de 10k noeuds par seconde mais le comportement de l'algorithmes était décevant car si le groupe est entouré de cases vides, il choisit un direction arbitraire rarement pertinente. <br>
  Les versions plus lourdes ont énormement réduit le nombre de noeuds calculables par seconde (~100) rendant l'exploration de l'arbre impossible.
  
  - Réduction du nombre de répartitions proposées à chaque groupe: <br>
  Au lieu de proposer à chaque groupe de se spliter en un, deux, trois, au plus sous-groupes à chaque tour, nous avons souhaiter réduire le nombre de possibilité (et ainsi le nombre de noeuds de l'arbre à parcourir). <br>
  Nous n'avons autoriser dans un premier temps que les divisions en deux sous-groupes, avec la contrainte qu'aucun des sous-groupes ne devait compter 1 seul monstre ou faire moins d'un tiers de la taille du groupe initiale. <br>
  Afin de gérer le nombre de répartitions pour les très gros groupes nous avons également choisis de ne conserver que les répartitions en sous-groupes d'environ 20%, 40%, 60% et 80%. <br>
  Enfin nous avons contraints tous les monstres de chaque groupe à bouger afin de ne plus avoir à considérer les mouvements où un sous-groupe restait immobile.
  
  - Limite du nombre de groupes considérés : <br>
  Afin de ne pas se retrouver bloquer face à un adversaire choississant sciemment de se spliter en plein de sous-groupes afin de générer un arbre des possibles trop grand pour répondre dans le temps imparti. Nous avons également limité le nombre de groupe considéré à chaque tour dans l'arbre. Nous considérons ainsi les plus grands groupes d'abord puis nous omettons les groupes restants.

  Ces différentes approches ont été implémentées dans l'algorithmes Algo_NegaMax_MPOO.py <br>
  
* #### Changement d'approche pour la génération de mouvements :
  Malheureusement même avec cette approche nous n'arrivions pas à réduire le facteur de branchement à moins de 30 sans perdre la plupart des mouvements intéressants. Nous avons donc revu la manière de réaliser le produit cartésien de possibilités des différents groupes. <br>
  - **Avant** : <br> 
  On limite pour chaque groupe le nombre de cases accessibles et de répartitions possibles. Puis on génère les mouvements possibles de la totalité des groupes en faisant le produit cartésien des posibilités <i>cases * réparitions</i> pour chaque groupes
  
  - **Après** : <br>
  On génère pour chaque groupe une liste de mouvements à partir de ses possibilités <i>cases * réparitions</i> puis on ordonne les mouvements de chaque groupe avec une nouvelle heuristiques (détaillée ci-dessous). Enfin on génère les mouvements de l'ensemble des groupes en considérant que les X premiers éléments du produit cartésien (des listes triées par notre heuristique). 




## <a name="heur"></a>4. Heuristiques

* #### Heuristique principale : Note de la carte (utilisée par tous les algos)
  L'heuristique nous permettant de noter les cartes de jeu ne donnant pas un des deux joueurs gagnants est <i>nombre\_alliés - nombre\_ennemis</i>. <br>
  Cette heuristique est appelée pour chaque noeuds à profondeur maximale d'un arbre. Sa simplicité est un avantage pour pouvoir parcourir l'arbre en profondeur.
  
* #### Heuristiques secondaires : (utilisées par certains algos)
  Ces heuristiques sont utilisées pour noter a priori des éléments dans la génération de mouvements
 
  - Heuristiques sur les cases entourant un groupe: (utilisée dans les algo MPOO)<br>
  L'heuristique "simple" finalement retenue pour noter les cases entourant un groupe de manière rapide compte très positivement les groupes d'humain mangeables, positivement les nombres d'ennemis mangeables, très négativement les groupes menaçant.
  
  - Heuristique sur les mouvements d'un groupe : (utilisée dans les algo MPO_2 et MPO_3)<br>
  Cette heuristique note un mouvement (un split ou non). Pour chaque case cible, il note la direction en fonction des groupes trouvés dans cette direction, de leur taille et de leur distance. Cela prend beaucoup plus de temps que de ne considérer que le contenu de la case mais cela permet de prendre en compte toute la carte. Afin d'accélerer un petit peu le temps de calcul nous avons rendu notre heuristique "myope" aux groupes éloignées de plus X cases.

Deux moyens pour comparer la performance des algorithmes:

* **Lancer un tournoi**

  Le fichier Tournoi dans le dossier Tournoi permet d'effectuer des duels entre différents algorithmes de décision (du dossier Algorithmes), sur plusieurs cartes données (du dossier Cartes).

* **Lancer une compétition "darwin" pour avoir une compréhension plus fine des hyper paramètres**

  Le darwin search prend en entrée un algorithme à hyper paramètres et génère une population en effectuant produit cartésien de ces paramètres discrétisés.
  C'est une variante de l'algorithme tournoi un peu spéciale car on va ajouter à un pool de combat aléatoirement ces individus, effectuer un tournoi et ne garder que les n meilleurs (les survivants).
  On répète cette idée pour voir quels algorithmes survivent le plus longtemps (ie restent dans le pool)
